## 2. NETTOYAGE DES DONN√âES
---

```python
# Cr√©er une copie pour le nettoyage
df_clean = df.copy()

# Nettoyage de la colonne 'rate'
def clean_rate(rate):
    if pd.isna(rate) or rate == 'NEW' or rate == '-':
        return np.nan
    try:
        rate_str = str(rate).split('/')[0].strip()
        return float(rate_str)
    except:
        return np.nan

df_clean['rate'] = df_clean['rate'].apply(clean_rate)
print(f"‚úì Notation moyenne : {df_clean['rate'].mean():.2f}/5")

# Nettoyage du co√ªt
def clean_cost(cost):
    if pd.isna(cost):
        return np.nan
    try:
        cost_str = str(cost).replace(',', '').strip()
        return float(cost_str)
    except:
        return np.nan

df_clean['approx_cost(for two people)'] = df_clean['approx_cost(for two people)'].apply(clean_cost)
print(f"‚úì Co√ªt moyen : {df_clean['approx_cost(for two people)'].mean():.0f} INR")

# Conversion des votes
df_clean['votes'] = pd.to_numeric(df_clean['votes'], errors='coerce')
print(f"‚úì Nombre moyen de votes : {df_clean['votes'].mean():.0f}")

# Suppression des doublons
if duplicates > 0:
    df_clean = df_clean.drop_duplicates()
    print(f"‚úì {duplicates} lignes dupliqu√©es supprim√©es")

print(f"\n‚úÖ Dataset nettoy√© : {df_clean.shape[0]} lignes √ó {df_clean.shape[1]} colonnes")
```
